```python
import pandas as pd
import tweepy
```

```python
#Leer las llaves 
keys = pd.read_csv("C:/Users/SoporteTI/OneDrive - Fundaci√≥n Universitaria Konrad Lorenz/nlp/twitter/desde_mi_balon.csv", header=None)
keys = dict(zip(keys[0], keys[1]))
keys.keys()
```

```python
#autenticar a trav√©s de la API

auth = tweepy.OAuthHandler(keys['api_key'], keys['api_secret_key'])
auth.set_access_token(keys['access_token'], keys['access_token_secret'])
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
#crear un tweet

#api.update_status('desde mi nevera')
```

```python
tweets = tweepy.Cursor(api.search,
                      q=[' -RT'],  #cualquier tema,
                      lang='es',
                      geocode='4.66734541166848,-74.06043916079582,2km', #Conocer de qu√© se twitea en una ubicaci√≥n,
                      tweet_mode = 'extended',
                      include_rts=False
                      ).items(200)
```

```python
#Objeto generador
t = [tweet for tweet in tweets]
```

```python
len(t)
```

```python
t[0].full_text
```

```python
t[0].created_at
```

```python
t[0].user.location
```

```python
t_processed = []


for tweet in t:
    temp = []
    temp.append(tweet.full_text)
    temp.append(tweet.created_at)
    temp.append(tweet.user.location)
    temp.append(tweet.id)
    temp.append(f"https://twitter.com/i/web/status/{tweet.id}")
    
    t_processed.append(temp)
```

```python
t_processed
```

```python
data = pd.DataFrame(t_processed)
data.columns = ['tweet', 'fecha', 'ubicaci√≥n', 'id', 'enlace']
data
```

```python
data['sentimiento_vader'] = data['tweet'].apply(lambda valor: vader.polarity_scores(valor)['compound'])
data.head()
```
